### Loading the dataset

```python
import pandas as pd
df = pd.read_csv("/content/lab1_dataset.csv")
print(df)
```

```python
df.shape
```

```python
t = df['Text'].values
l = df['Label'].values
print(df.head())
```

### Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
v = CountVectorizer()

v1 = TfidfVectorizer()

f = v.fit_transform(t)
f1 = v1.fit_transform(t)
print(f1)
```

### MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
ml = MLPClassifier()
ml.fit(f1,l)
```

### Testing

```python
text_t = ['The movie is not bad']
text_t1 = v1.transform(text_t)
print(text_t1)
```

```python
output = ml.predict(text_t1)
print(output)
```

### SVM Classifier

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(f1,l)
```

```python
svm_output = sv.predict(text_t1)
print(svm_output)
```

### Naive Baiyes

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(f1,l)
```

```python
nb_output = nb.predict(text_t1)
print(nb_output)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(f1,l)
```

```python
rf_output = rf.predict(text_t1)
print(rf_output)
```
